## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras import layers
from tensorflow.keras.layers import Input, Dense, Dropout
from sklearn.preprocessing import OneHotEncoder
#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_attrition_df = attrition_df[['Attrition', 'Department']]
y_attrition_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
selected_columns = ['Age'
                  , 'HourlyRate'
                  , 'YearsAtCompany'
                  , 'JobLevel'
                  , 'JobSatisfaction'
                  , 'PerformanceRating'
                  , 'Education'
                  , 'WorkLifeBalance'
                  , 'DistanceFromHome'
                  , 'YearsSinceLastPromotion']



# Create X_df using your selected columns
X_attrition_df = attrition_df[selected_columns]

# Show the data types for X_df
X_attrition_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_attrition_df, y_attrition_df, test_size=0.2, random_state=42)
```

```python
# Convert your X data to numeric data types however you see fit
from sklearn.pipeline import Pipeline # Import the Pipeline class
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

numeric_features = selected_columns
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])
# Add new code cells as necessary
from sklearn.compose import ColumnTransformer
# Import ColumnTransformer if it hasn't been already
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features)])
```

```python
# Create a StandardScaler

numeric_features = selected_columns
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

```

```python
X_train = pd.DataFrame(X_train, columns=selected_columns)
X_test = pd.DataFrame(X_test, columns=selected_columns)

X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)

# X_train = preprocessor.fit_transform(X_train)
# X_test = preprocessor.transform(X_test)
```

```python
# Create a OneHotEncoder for the Department column
department_encoder = OneHotEncoder(sparse=False)
y_train_department = department_encoder.fit_transform(y_train[['Department']])
y_test_department = department_encoder.transform(y_test[['Department']])

# Fit the encoder to the training data


# Create two new variables by applying the encoder
# to the training and testing data
print("X_train sample:")
print(X_train[:5])
print("y_train_department sample:")
print(y_train_department[:5])


```

```python
# Create a OneHotEncoder for the Attrition column
attrition_encoder = OneHotEncoder(sparse=False)
y_train_attrition = attrition_encoder.fit_transform(y_train[['Attrition']])
y_test_attrition = attrition_encoder.transform(y_test[['Attrition']])

# Fit the encoder to the training data
print("X_train sample:")
print(X_train[:5])

# Create two new variables by applying the encoder
# to the training and testing data
print("y_train_attrition sample:")
print(y_train_attrition[:5])
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
input_shape = X_train.shape[1]

# Create the input layer
input_layer = Input(shape=(input_shape,))

# Create at least two shared layers
shared_layer_1 = Dense(64, activation='relu')(input_layer)
shared_layer_2 = Dense(32, activation='relu')(shared_layer_1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer
department_hidden_layer = Dense(16, activation='relu')(shared_layer_2)
department_output_layer = Dense(y_train_department.shape[1], activation='softmax', name='department_output')(department_hidden_layer)

# Create the hidden layer


# Create the output layer

```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer
attrition_hidden_layer = Dense(16, activation='relu')(shared_layer_2)
attrition_output_layer = Dense(y_train_attrition.shape[1], activation='softmax', name='attrition_output')(attrition_hidden_layer)

# Create the hidden layer


# Create the output layer

```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[department_output_layer, attrition_output_layer])

# Compile the model
model.compile(optimizer='adam',
              loss={'department_output': 'categorical_crossentropy', 'attrition_output': 'categorical_crossentropy'},
              metrics={'department_output': 'accuracy', 'attrition_output': 'accuracy'})


# Summarize the model
model.summary()
```

```python
# Train the model
history = model.fit(X_train, {'department_output': y_train_department, 'attrition_output': y_train_attrition},
                    epochs=50, batch_size=32, validation_split=0.2)
```

```python
# Evaluate the model with the testing data
evaluation = model.evaluate(X_test, {'department_output': y_test_department, 'attrition_output': y_test_attrition})
```

1. Accuracy is a commonly used metric for classification problems; however, it may not always be the best metric, especially if the dataset is imbalanced. In this project, we are predicting employee attrition and department, both of which might have class imbalances (e.g., fewer people leaving the company than staying, or some departments having significantly more employees than others).

For imbalanced datasets, other metrics such as precision, recall, F1-score, or area under the ROC curve (AUC) might provide better insights into the model's performance. These metrics can help evaluate how well the model handles different classes and avoid misleading conclusions that might arise from using accuracy alone.

2. For both output layers (department and attrition), I chose the softmax activation function. The reasons are as follows:

Softmax for Department Prediction: The department prediction is a multi-class classification problem, where we need to predict one out of several possible departments. The softmax activation function is suitable for this purpose as it outputs a probability distribution across all classes, and the sum of these probabilities equals one. This allows us to select the department with the highest probability as the predicted class.

Softmax for Attrition Prediction: Similarly, although attrition might seem like a binary classification problem, we treated it as a categorical problem with two classes (stay or leave). The softmax activation function ensures that the output provides the probability of each class, making it easy to interpret and handle in a similar manner as the department prediction.
